```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Impoerting the dataset file
titanic_data = pd.read_csv('/content/drive/MyDrive/CodeSoft/Task 1/Titanic-Dataset.csv')
```

```python
# Giving shape and features to the dataset
shape = titanic_data.shape
features = titanic_data.columns
print(f"Our data has the following shape {shape}")
print(f"Features present in our data are:",features.to_list(),sep="\n")
```

```python
plt.pie(
    titanic_data['Survived'].value_counts().sort_values(ascending=False),
    labels = ['Died','Survived'],
    colors=["red", "Green"],
    autopct="%1.1f%%",
)

#here is the adding of legend to pie chart
plt.legend(["Died", "Survived"], loc="upper right")

plt.title('Survival Status of the dataset')
plt.show()
```
